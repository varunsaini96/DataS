Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
import plotly.graph_objects as go
import plotly.express as plx
```

```python
plx.scatter(x = titanic.age, y = titanic.fare)
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips = sns.load_dataset("tips")
tips
```

```python
import plotly.express as plx
```

```python
plx.box(tips, x = tips.day, y = tips.total_bill)
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in

Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans. Distplot:- A distplot is short of distribution plot, it is a graphical representation used to visualize the distribution of a single-variable dataset. It combines elements of a histogram and a kernel density plot to provide insights into the underlying distribution of the data

```python
plx.histogram(tips, x=tips.total_bill, y = tips.tip, color = "sex", marginal="rug")
```

```python
def re_index(df1):
    df1 = df1.reset_index(drop=True)
    new_index = range(1, len(df1)*2, 2)
    df1.index = new_index
    return df1
```

```python
def sum_of_values(df):
    try :
        first_three_values = df['Values'].head(3)
    
        sum_of_first_three_values = first_three_values.sum()
    
    except Exception as e:
        print("You should take column as Values only :",e)
        
    else:
        print(f"The sum of the first three values is: {sum_of_first_three_values}")
```

```python
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
```

```python
df
```

```python
sum_of_values(df)
```

```python
def count_words(df):
    count_words = lambda text: len(text.split())
    df['Word_Count'] = df['Text'].apply(count_words)
    
    return df
```

```python
data = {'Text': ['Hello there', 'I hope everyone is doing great', 'Stay happy and hard working']}
df = pd.DataFrame(data)

count_words(df)
```

```python
def cal(df) :
    mean = df["Values"].mean()
    median = df["Values"].median()
    mode = df["Values"].mode()
    print("Mean:",mean)
    print("Median:",median)
    print("Mode:",mode)
```

```python
def calculate_moving_average(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date')
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df
```

```python
df = pd.DataFrame({'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05', '2023-09-06', '2023-09-07'],
                   'Sales': [100, 200, 150, 180, 250, 300, 220]})
```

```python
calculate_moving_average(df)
```

```python
def create_weekday(df) :
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df
```

```python
df = pd.DataFrame({"Date" :  ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"]})
```

```python
create_weekday(df)
```

```python
def Exception_hierarchy(exception_class):
    print(exception_class.__name__)
    
    for i in exception_class.__subclasses__():
        Exception_hierarchy(i)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is a base class for exceptions related to looking up values in sequences, mappings, or other data structures. 

```python
# KeyError:
# This error occurs when trying to access a dictionary with a key that doesn't exist in the dictionary.

dct = {"apple": 3, "banana": 2, "orange": 5}

try:
    value = dct["grape"]
except KeyError as k:
    print("There is an error:", k)
```

```python
# IndexError:
# This error occurs when trying to access an index in a sequence that is outside the valid range of indices.

lst = [10, 20, 30, 40, 50]

try:
    value = lst[10]
except IndexError as i:
    print("There is an error:", i)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError:
ImportError is a base class for exceptions related to importing modules. It can be raised for various reasons, including when a requested module cannot be found, when there's a problem with the module's contents, or when there's an issue with circular imports.

```python
# ModuleNotFoundError:- you do not have the module you tried importing installed on your computer.

try :
    import pwskills
    
    pwskills.append()
    
except ModuleNotFoundError as M :
    print("There is an error:", M)
```

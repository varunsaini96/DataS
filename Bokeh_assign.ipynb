Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting

from bokeh.plotting import figure, output_file, show

x = [2,2,8,4,5]
y = [3,4,5,10,7]

output_file("line.html")
p = figure(title = "line graph")
p.xaxis.axis_label = "X line"
p.yaxis.axis_label = "Y line"
p.scatter(x,y, size = 10, fill_color = "red")
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans. Glyphs are visual representations of data points or geometrical shapes that can be added to a plot to visualize your data. Glyphs are the fundamental building blocks of Bokeh plots, and they allow you to create various types of visualizations, such as lines, circles, squares, and more. You can add glyphs to a Bokeh plot to represent your data in a specific way.

```python
x = [1,2,3,4,5]
y = [3,4,5,6,7]

output_file("line.html")
p = figure(title = "line graph")
p.xaxis.axis_label = "X line"
p.yaxis.axis_label = "Y line"
p.square(x,y, size = 10, fill_color = "pink") #Glyphs used over here as an square
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Bokeh Server is a component of the Bokeh library that allows you to create interactive web applications and dashboards with live, real-time updates. It enables you to build dynamic and data-driven web applications with Bokeh plots and widgets that respond to user input or changes in data without requiring constant re-rendering or page refreshes. Bokeh Server is particularly useful when you need to create interactive data visualizations or dashboards that require live updates, interactions, and user feedback.

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import Slider
```

```python
x = [4, 6, 7, 7, 5]
y = [3, 4, 5, 6, 7]


output_file("line.html")


p = figure(title="Line Graph")
p.xaxis.axis_label = "X line"
p.yaxis.axis_label = "Y line"


line = p.line(x, y, line_width=2, line_color="green")


slider = Slider(start=0, end=10, value=3, step=1, title="Slider")

# Define a callback function for the slider
def update_slider(attr, old_value, new_value):
    line.data_source.data['y'] = [i * new_value for i in [1, 2, 3, 4, 5]]

slider.on_change('value', update_slider)

show(p)
```
